```python
dats = pd.read_csv("ctp6.csv", header = 0)
dfTrain, dfTest = train_test_split(dats, test_size=0.2)

dfTrainVals = dfTrain["phase"]
dfTrain = dfTrain.drop("phase",1)

dfTestVals = dfTest["phase"]
dfTest = dfTest.drop("phase",1)

print("DONE - SPLIT")
```

```python
from catboost import CatBoostRegressor
dfTrain.fillna('', inplace=True)
catTypes = np.where(dfTrain.dtypes != float)[0]
model_cb = CatBoostRegressor(cat_features = catTypes)
model_cb.fit(dfTrain, dfTrainVals, eval_set = (dfTest, dfTestVals), plot = True)

print("DONE - MODEL FIT")
```

```python
dfFeatImp = pd.DataFrame(model_cb.get_feature_importance(prettified=True))
#plotting feature importance
plt.figure(figsize=(12, 6));
featPlt= sns.barplot(x="Importances", y="Feature Id", data=dfFeatImp,palette="cool");

print("DONE - FEATURE IMPORTANCE")
```
